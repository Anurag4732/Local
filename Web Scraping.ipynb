## Multiprocessing and Multithreading

```python
import threading
import time

def sqr(x):
    time.sleep(0.2)
    return([i*i for i in x])
    

def qub(x):
    time.sleep(0.2)
    return([i*i*i for i in x])

# t1 = time.time()
# print(sqr([1,2,3]))  # 0.4100
# print(qub([1,2,3]))
# t2 = time.time()
# print(t2-t1)
t1 = time.time()
l = [1, 2, 3]
thred1 = threading.Thread(target = sqr, args = (l))
thred2 = threading.Thread(target = qub, args = (l))

thred1.start()
thred2.start()

thred1.join()
thred2.join()

t2 = time.time()
print(t2-t1)
```
